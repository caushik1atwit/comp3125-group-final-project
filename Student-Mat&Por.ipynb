```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def load_data(file_path):
    return pd.read_csv(file_path, delimiter=';')

df_math = load_data("student-mat.csv")
df_por = load_data("student-por.csv")

drop_columns = ['studytime', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'schoolsup', 'famsup', 'paid', 'famrel', 'goout', 'Dalc', 'Walc', 'G1', 'G2', 'G3']
df_math = df_math.drop(columns=drop_columns)
df_por = df_por.drop(columns=drop_columns)
```

```python
def cap_outliers(df, column, cap_value):
    df[column] = df[column].apply(lambda x: min(x, cap_value))
    return df

df_math = cap_outliers(df_math, 'absences', 30)
df_por = cap_outliers(df_por, 'absences', 30)
```

```python
def display_cleaned_data(df, num_rows=10):
    from IPython.display import display
    display(df.head(num_rows))

display_cleaned_data(df_math)
display_cleaned_data(df_por)
```
